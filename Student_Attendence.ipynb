```python
df=df.drop_duplicates(keep="first")
data_duplicated=df.duplicated().sum()
print("the duplicated data present:",data_duplicated)
```

```python
df=df.drop(columns=["Topic","Type","Host name","Host email","Department","Group","Source","Max concurrent views","Guest","In waiting room","Creation time","Unique viewers","Recording disclaimer response","ID"])
```

```python
df.head(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Select the numeric columns
numeric_cols = ['Duration (minutes)', 'Total participant minutes', 'Duration (minutes).1']

# Scale values (optional, helps if ranges differ a lot)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numeric_cols])

# Compute correlation matrix
corr_matrix = np.corrcoef(scaled_data.T)

# Plot heatmap with values
fig, ax = plt.subplots(figsize=(6,6))
cax = ax.matshow(corr_matrix, cmap='coolwarm')

# Add colorbar
plt.colorbar(cax)

# Set ticks and labels
ax.set_xticks(range(len(numeric_cols)))
ax.set_yticks(range(len(numeric_cols)))
ax.set_xticklabels(numeric_cols, rotation=45, ha='left')
ax.set_yticklabels(numeric_cols)

# Overlay correlation values on the heatmap
for i in range(len(numeric_cols)):
    for j in range(len(numeric_cols)):
        text = ax.text(j, i, f"{corr_matrix[i, j]:.2f}",
                       ha="center", va="center", color="black")

plt.title("Correlation Heatmap of Attendance Features", pad=20)
plt.show()
```

```python
# Calculate attendance percentage
df['Attendance_Percentage'] = (df['Duration (minutes).1'] / 120) * 100

# Apply eligibility rule (>= 80%)
df['Eligibility'] = np.where(df['Attendance_Percentage'] >= 80, 'Eligible', 'Not Eligible')

# Summary counts
print(df['Eligibility'].value_counts())

# Plot histogram of attendance percentages
plt.figure(figsize=(8,5))
plt.hist(df['Attendance_Percentage'], bins=20, color='skyblue', edgecolor='black')
plt.axvline(80, color='red', linestyle='--', label='80% Threshold')
plt.title("Distribution of Attendance Percentage")
plt.xlabel("Attendance %")
plt.ylabel("Number of Participants")
plt.legend()
plt.show()

# Bar chart of eligibility counts
plt.figure(figsize=(6,4))
df['Eligibility'].value_counts().plot(kind='bar', color=['green','orange'])
plt.title("Certificate Eligibility")
plt.xlabel("Status")
plt.ylabel("Number of Participants")
plt.show()
```
